```python
import gensim
from gensim.models import Word2Vec
```

```python
df2 = df1.copy()
```

```python
df2['reviews'] = df2.reviews.apply(gensim.utils.simple_preprocess)
df2.reviews[0]
```

```python
df2.reviews[2]
```

```python
df2.head()
```

```python
model = Word2Vec(
    window = 5,
    min_count= 3,
    workers = 4
)
```

```python
model.build_vocab(df2.reviews)
```

```python
model.corpus_count
```

```python
model.wv.most_similar("phone")
```

```python
model.train(df2.reviews, total_examples=model.corpus_count, epochs = 10)
```

```python
# import multiprocessing
# multiprocessing.cpu_count()  ---> This tells the no. of workers to be used in word2vec model
```

```python
if 'phone' in model.wv:
    print("hi")
```

```python
df3 = df2.copy()
```

```python
import numpy as np

def get_review_vector(review, model):
    vectors = [model.wv[word] for word in review if word in model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)
```

```python
df3['vectors'] = df3.reviews.apply(lambda x: get_review_vector(x,model))
```

```python
df3.head()
```

```python
df3.vectors[0]
```

# Balancing the dataset

```python
df4 = df3.copy()
```

```python
df4.head()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sampler = SMOTE(random_state = 42)
```

```python
df4 = df4.drop("reviews",axis = 1)
df4.head()
```

```python
X = np.vstack(df4["vectors"].values)
y = df4.rating
```

```python
x_samp, y_samp= sampler.fit_resample(X,y)
```

```python
(X.shape,y.shape),(x_samp.shape,y_samp.shape)
```

```python
y_samp.value_counts()
```

```python
y_samp.isna().sum()
```

```python
df4.isna().sum()
```

```python
x_samp[0]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
(x_samp.shape,y_samp.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_samp,y_samp,test_size = 0.2, random_state = 42)
```

# Using KNN:

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
knn_pred = knn.predict(x_test)
print(classification_report(y_test,knn_pred))
```
